```python
for num in range(1,31):
    df_two=pd.read_csv('./df_total_{}.csv'.format(num))
    df_two['ts'] = pd.to_datetime(df_two['ts'], format='%Y_%m_%d_%I_%M')

    # 날짜별(00:00:00 ~ 23:59:59) 인덱스 컬럼 생성
    ts_group = []
    for i in range(df_two.shape[0]):
        ts_group.append(df_two['ts'].iloc[i].strftime("%Y-%m-%d"))
    df_two['ts_group'] = ts_group

    # Sleep (수면 중)인 데이터 삭제
    df_two = df_two[df_two['action']!='sleep']

    # 날짜 활동별 시간(분 단위) 나타내는 데이터프레임 생성
    df = pd.DataFrame({i:[] for i in df_two['action'].unique()})
    from tqdm import tqdm

    for i in df_two['ts_group'].unique():
        df_sample = df_two[df_two['ts_group'] == i]
        dic_total = {i : 0 for i in df_two['action'].unique()}
        
        for j in range(df_sample.shape[0]-1):
            td = df_sample['ts'].iloc[j+1] - df_sample['ts'].iloc[j]
            col = df_sample.iloc[j]['action']
            dic_total[col] += td.total_seconds()/60
            
        df = df.append(dic_total, ignore_index=True)
    df['ts_group'] = df_two['ts_group'].unique()
    df['Month']=df.ts_group.str.split('-').str[1]
    df['Day']=df.ts_group.str.split('-').str[2]
    df.drop(['ts_group'],axis=1,inplace=True)
    df.to_csv('./user{}_worktime.csv'.format(num))
```
